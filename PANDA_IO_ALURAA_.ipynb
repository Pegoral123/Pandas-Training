# obtendo arquivos JSON de ***APIs***

```python
import requests
import json
```

```python
url = requests.get('https://jsonplaceholder.typicode.com/users')
dados_user =json.loads(url.text) #lendo o link o arquivo em formato json e transformado ele em um dataframe para que possa analisar de forma melhor
dados_normalizados = pd.json_normalize(dados_user, sep='_')  #argumento sep='_' no método pd.json_normalize, você garante que os subcampos de dados aninhados sejam desaninhados com os níveis unidos por um sublinhado (_).
dados_normalizados.head()
```
