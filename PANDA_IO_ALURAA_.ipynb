# **MANIPULANDO ARQUIVO XLSX**

```python
url = 'https://github.com/alura-cursos/Pandas/blob/main/emissoes_CO2.xlsx?raw=True' #raw = True , ele traz o arquivo no formato bruto
dados = pd.read_excel(url) #read_excel abrindo arquivo do formato xlsx
dados.head()
```

```python
pd.ExcelFile(url).sheet_names # metodo ExcelFile manipula o arquivo excel e metodo sheet_names que dar retorno dos nomes das paginas disponivel na planilha do excel que foi informado
```

```python
percapita = pd.read_excel(url, sheet_name = 'emissoes_percapita') #aqui especificamos no metodo sheet_name que vamos querer verificar a pagina "emissoes_percapita"
```

```python
percapita.head()
```

```python
fontes = pd.read_excel(url, sheet_name = 'fontes')
fontes.head()
```

```python
emissoes_C02 = pd.read_excel(url, sheet_name = 'emissoes_C02', usecols='A:D', nrows= 10) #metodo usecols nos permite escolhar quantidade de colunas que queremos mostrar e nrwos mostra a quantidade de linha que queremos vê
emissoes_C02.head()
```

```python
percapita.to_excel('c02_percapita.xlsx', index=False) #salvando o arquivo no formato xlsx
```
